```python
import pandas as pd
```

```python
s = pd.Series([10, 20, 30, 40, 50])
s.name = "calories"
s
```

```python
s[0]
```

```python
s[0:3]
```

```python
# ilocc -> location based indexing
```

```python
s.iloc[3]
```

```python
s.iloc[[1,3,4]]
```

```python
index = ["apple","banana","graps","orange","pine-apple"]
s.index = index
s
```

```python
s['graps']
```

```python
# s.iloc["graps"] # -> only used for numerical index
```

```python
# loc -> label bases indexing 
```

```python
s.loc["graps"]
```

# in label based indexing both start and end values are included in the output

```python
s.loc["banana":"pine-apple"]
```

```python
s.loc[["banana","pine-apple"]]
```

```python
fruit_protein = {
    "Avocado": 2.0,      # grams of protein
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name = "Protein")
s2
```

## Conditional indexing 

```python
s2>1
```

```python
s2[s2>1]
```

```python
# logical operators ( and & or | )
```

```python
s2[(s2>0.5) & (s2<2)]
```

```python
s2[(s2>0.5) | (s2<2)]
```

```python
# not operation
s2[~(s2>1)]
```
