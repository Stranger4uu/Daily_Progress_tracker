```python
# np.zeros -> this will create arrays full of zeros
array = np.zeros([2,5]) # 2 rows and 5 columns 
array
```

```python
array = np.ones([2,4]) #array full of one
array
# default it gives float value but we can change it's datatype too
arr = np.ones([2,3], dtype = int)
arr
```

```python
# np.full create an array full of any value
array = np.full([2,5], 4) #row , column , and default value
array
```

```python
# np.empty -> uninitialized array -> create an array without setting any value
"""
ðŸš€ Why does np.empty() exist then?

Because it is:

Faster than zeros() or ones()

Useful when you are going to overwrite all values immediately
"""
array1 = np.empty([2,3])
array1
```
