```python
# in py we use and & or but in numpy we use ligical_and & logical_or
```

```python
# logical_and
arr = np.array([10,30,45,35,96,56])
mask = np.logical_and(arr > 15 , arr <96)
print(mask)
arr[mask]
```

```python
arr2 = np.array([10,30,45,35,96,56])
mask2 = np.logical_or(arr > 30 , arr <96)
print(mask2)
arr2[mask2]
```

```python
# no zero
arr = np.array([0,4,0,5,6,0,34,3,6,0])
non_zero = np.nonzero(arr)
print(non_zero)
```
