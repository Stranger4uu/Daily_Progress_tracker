```python
import numpy as np
```

```python
# np.nan -> not a number
```

```python
array = np.array([1,2,4,np.nan,5,8])
print(array)
```

```python
# np.inf and -np.inf -> positive and negative infinities
array2 = np.array([1,2,4,np.inf,5,8])
print(array2)
```

```python
# np.isinf
# np.isnan
# np.isfinite

#-> these are the functons ised to detect any infinite value, null value and finite value
```

```python
print(np.isnan(array))
```

```python
print(np.isinf(array2))
```

```python
print(np.isfinite(array))
```
