## Swapaxes -> swap 2 specific axes in a matrix

```python
arr = np.array([[[1,3],[4,5]]])
arr.shape
```

```python
swap = arr
swap.shape
```

```python
arr = np.swapaxes(arr, 0,1)
swap.shape
```
