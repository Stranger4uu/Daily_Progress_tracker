Vectorization means performing operations on entire arrays at once, instead of using Python for loops.

In simple words:

âŒ Loop â†’ element by element
âœ… Vectorization â†’ whole array in one shot

NumPy is fast because it uses optimized C code internally, so vectorized operations are much faster than Python loops.

âš¡ Much faster execution

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
result = []
for i in range(len(a)):
    result.append(a[i] + b[i])
print(result)
```

This is slow because:
- Python loop runs every iteration
- Each operation happens separately

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

result = a + b
print(result)
```

NumPy performs addition at low-level (C implementation) â†’ much faster.

```python
# broadcasting and Vectorization
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([10, 20, 30])

print(A + B)
```

```python
# np.vectorize -> convert a regular function to be applied on an array
```

```python
def square(x):
    return x * x


vfun = np.vectorize(square)
```
