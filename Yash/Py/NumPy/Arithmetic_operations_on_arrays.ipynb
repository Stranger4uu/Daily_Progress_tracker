```python
import numpy as np
```

```python
a = np.array([1,4,6])
b = np.array([4,6,1])
```

```python
# addition
print(a+b)
```

```python
# substraction
print(a-b)
print(b-a)
```

```python
# division 
print(a/b)
print(a//b) # -> integers value
print(b//a) # -> integers value
```

```python
# multipliacation
print(a*b)
```

```python
# modulus
print(a%b)
```

```python
# exponent
print(a**b) # -> power
print(a**2)
```

### Universe functions -> Ufuncs

```python
arr = np.array([1,4,9,16])
```

```python
# sqrt (square root)
print(np.sqrt(arr))
```

```python
# exponentioal -> np.exp -> e^x , where x is any integer
print(np.exp([2,3]))
```

```python
# sine function -> np.sin
angles = np.array([0,np.pi,np.pi/2])
print(np.sin(angles))
```
