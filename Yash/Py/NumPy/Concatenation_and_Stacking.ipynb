```python
import numpy as np
```

```python
a = np.array([2,4])
b = np.array([3,6])

combine = np.concatenate((a,b))
```

```python
print(combine)
```

```python
# Vstack -> verticle stack
a = np.array([[2,4],[1,2]])
b = np.array([[3,6],[2,8]])
stack = np.vstack((a,b))
print(stack)
```

```python
# Hstack -> Horizontal stack
a = np.array([[2,4],[1,2]])
b = np.array([[3,6],[2,8]])
stack = np.hstack((a,b))
print(stack)
```

```python
# Stack -> ((a,b), axis = )
# axis 0 = row wise
# axis 1 = column wise
a = np.array([[2,4],[1,2]])
b = np.array([[3,6],[2,8]])
stack = np.stack((a,b),axis = 0)
print(stack)
```

```python
a = np.array([[2,4],[1,2]])
b = np.array([[3,6],[2,8]])
stack = np.stack((a,b),axis = 1)
print(stack)
```
