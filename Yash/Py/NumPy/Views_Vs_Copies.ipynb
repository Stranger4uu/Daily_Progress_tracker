```python
import numpy as np
```

# whenever we create a view our original arrays going to effected but
# whenever we create a copy our original arrays will not be effected and it will be uncahnged and our copy will be effected

```python
# view 
array = np.array([1,2,3,4,5,6,7])
view = array[1:5]
view
```

```python
view[0] = 69
view
```

```python
array
```

```python
# so original array is also changed here 
```

```python
# copy
copy = array[1:3].copy()
copy
```

```python
copy[0] = 59
copy
```

```python
array
```

```python
copy = array.copy()
copy
```

```python
copy[3] = 79
copy
```

```python
array
```

```python
# so here original array is not effected at all
```
