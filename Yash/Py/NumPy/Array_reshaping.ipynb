```python
import numpy as np
# Reshape
arr = np.array([1,2,3,4,5,6])
arr
```

```python
reshaped = arr.reshape(3,2)
print(reshaped)
```

```python
reshaped2 = reshaped.reshape(2,3)
print(reshaped2)
```

```python
# ravel -> convert to 1D array
# it returns a view
ravel = reshaped2.ravel()
ravel
```

```python
# flatten -> also convert to 1D array
# return a copy of an array 
flat = reshaped2.flatten()
flat
```

```python
ravel[0] = 69
print(ravel)
print(reshaped2)
```
